#### train_test_split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=120)

```

```python
print(y_train.shape)
y_test.shape
```

###  cross validation

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
log_score=cross_val_score(LogisticRegression(),x_train,y_train)
svm_score=cross_val_score(SVC(),x_train,y_train)
rf_score=cross_val_score(RandomForestClassifier(),x_train,y_train)
```

```python
import numpy as np
print(np.mean(log_score))
print(np.mean(svm_score))
print(np.mean(rf_score))
```

### choosing model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

###  train model

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

### 5 evaluate model

```python
y_pred=model.predict(x_test)
y_pred
```
